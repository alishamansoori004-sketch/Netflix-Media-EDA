```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

# Display plots inside the notebook
%matplotlib inline
```

```python
import pandas as pd

df = pd.read_csv("../data/netflix_titles.csv")

df.head()
```

```python
# Shape of the dataset
print("Shape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types
print("\nData Types:")
print(df.dtypes)

# First 5 rows
df.head()
```

```python
# General information
df.info()

# Statistical summary
df.describe(include='all')
```

```python
# Count missing values
missing_values = df.isnull().sum()

print(missing_values)
```

```python
# Missing value percentage
missing_percentage = (df.isnull().sum() / len(df)) * 100

print(missing_percentage.round(2))
```

```python
# Duplicate rows
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
# Fill missing values with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Fill rating with the most frequent value (mode)
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
print(df.isnull().sum())
```

```python
type_counts = df['type'].value_counts()

print(type_counts)
```

```python
plt.figure(figsize=(6,4))

type_counts.plot(kind='bar')

plt.title("Distribution of Movies and TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

# Observation 1: Distribution of Movies and TV Shows

### Observation
- Netflix contains 6,131 Movies and 2,676 TV Shows.
- Movies account for approximately 70% of the total content.
- TV Shows account for approximately 30% of the total content.

### Business Insight
- Netflix's catalog is dominated by movies.
- Investing in quality TV Shows may increase user engagement and retention.

```python
plt.figure(figsize=(6,4))

type_counts.plot(kind='bar')

plt.title("Distribution of Movies and TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.tight_layout()

plt.savefig("../images/movies_vs_tvshows.png")

plt.show()
```

```python
release_year_counts = (
    df['release_year']
    .value_counts()
    .sort_index()
)

print(release_year_counts.tail())
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    release_year_counts.index,
    release_year_counts.values,
    linewidth=2
)

plt.title("Netflix Content Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.show()
```

# Observation 2: Netflix Content Released Over the Years

### Observation
- The number of Netflix titles released increased gradually over the years.
- A significant rise is observed after 2015.
- The highest number of titles was released between 2018 and 2020.
- After 2020, the number of releases shows a slight decline.

### Business Insight
- Netflix expanded its content library rapidly during 2015–2020 to attract more subscribers.
- The decline after 2020 may be due to changes in content strategy and the impact of the COVID-19 pandemic on production.
- Understanding yearly release trends helps Netflix plan future content investments and maintain audience engagement.

```python
plt.figure(figsize=(14,6))

plt.plot(
    release_year_counts.index,
    release_year_counts.values,
    linewidth=2
)

plt.title("Netflix Content Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.grid(True)

plt.tight_layout()

plt.savefig("../images/release_year_trend.png")

plt.show()
```

```python
# Select only Movies
movies = df[df['type'] == 'Movie'].copy()

# Remove rows where duration is missing
movies = movies.dropna(subset=['duration'])

# Extract numeric duration
movies['duration_minutes'] = (
    movies['duration']
    .str.extract('(\d+)')[0]
    .astype(int)
)

movies.head()
```

```python
plt.figure(figsize=(10,6))

plt.hist(movies['duration_minutes'], bins=20)

plt.title("Distribution of Movie Duration")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")

plt.tight_layout()

plt.savefig("../images/movie_duration.png")

plt.show()
```

# Observation 8: Movie Duration Distribution

### Observation
- Most Netflix movies have a duration between 80 and 120 minutes.
- Very short and very long movies are less common.
- The distribution indicates that standard-length feature films dominate the platform.

### Business Insight
- Movies with a duration of around 90–120 minutes are likely to match viewer preferences.
- Understanding duration trends helps Netflix plan future content that aligns with audience viewing habits.

```python
# Remove leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()

# Convert to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Create year_added column
df['year_added'] = df['date_added'].dt.year

# Count content added each year
content_added = df['year_added'].value_counts().sort_index()

print(content_added)
```

```python
plt.figure(figsize=(12,6))

plt.plot(content_added.index, content_added.values, marker='o')

plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")

plt.grid(True)
plt.tight_layout()

plt.savefig("../images/content_added_by_year.png")

plt.show()
```

# Observation 9: Content Added to Netflix by Year

### Observation
- Netflix added relatively few titles before 2015.
- A rapid increase in content additions is observed from 2016 onwards.
- The highest number of titles was added around 2019–2020.
- After 2020, content additions show a slight decline.

### Business Insight
- Netflix aggressively expanded its library between 2016 and 2020 to strengthen its market position.
- The slowdown after 2020 may reflect strategic content planning and production challenges.
- Monitoring yearly content additions helps Netflix optimize investment in new titles.
